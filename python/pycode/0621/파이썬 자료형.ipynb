### 파이썬 자료형
- 숫자, 문자, 리스트, 듀플, 딕셔너리
#### 숫자



```python
int1=123
int2=12.
int3=int1+int2    # 숫자형 + 숫자형  => 산술연산 더하기

print(type(int1), type(int2),type(int3))   # type(): 데이터 형식을 출력
print(int3)
```

#### 문자형
- 데이터 입력시 인용부호로 단일(') 또는 이중(") 따움표를 사용한다.
- 문자열은 문자들의 모임, 문자를 나열한 것을 문자열이라고 함
- 인덱싱(indexing)/슬라이싱



```python
str1="Python is Very Powerful"
str2='파이썬, 문자열, 데이터'
str3="12345"

print(type(str1),type(str2), type(str3))
print(str2 + str3)  # 문자 + 문자  => 연결연산
```

```python
# 문자데이터 인덱싱
print(str1[0])
print(str1[-1])
print(str2[4])

# 문자데이터 슬라이싱
print(str2[0:3])  #  str2[:3]
print(len(str1))  # len() : 입력된 데이터 길이
print(str1[-8:]) # print(str1[-8:23])
```

```python
print(str2.split(", "))   # split() 입력 기호를 중심으로 텍스트 나누기, 기본값은 스페이스
print(str2)
```

```python
print(str1.replace("Python", "파이썬"))  # replace("a", "b") : a값을 찾아서 b로 변경
```

```python
str3="abcd"
"-".join(str3)
```

#### 리스트형
- 데이터 입려시 [ ] 로 묶어서 입력
- 리스트명 =[요소1, 요소2, 요소3, ...]
- 복합형 데이터 입력시 사용
- 인덱싱/슬라이싱



```python
lst1=[10,20,30,'파이썬', "삼성멀티캠", [150,300,500]]
lst2=['요소1', "요소2"]

print(type(lst1), type(lst2))
```

```python
print(lst1[0])
print(lst1[-1])
print(lst1[-1][1])
```

```python
print(lst1[:3])
print(lst1[4])
print(type(lst1[4]))
print(lst1[4][0:2])
```

```python
lst1+lst2
```

```python
lst2.append('서정동')  # 리스트 구조의 마지막 데이터 뒤에 추가
lst2
```

```python
lst1[3]='파이썬 3.9'
lst1
```

```python
lst1[-2]=lst1[-2].replace('캠', "캠퍼스")
lst1
```

```python
print(len(lst1))   # 리스트의 데이터 개수
print(len(lst1[4]))  # lst1의 4번째 인덱스의 값 길이
```

#### 튜플형
- 입력시 ()로 묶어서 입력
- 튜플변수=(10,20,30)
- 리스트의 항목 값은 변화가 가능하고 튜플의 항목 값은 변화가 불가능하다.

```python
t1=(1, 2, "파이썬", (100,200))

print(type(t1))
print(t1[:3])
print(t1[2])
```

```python
t1[2]="파이썬 3.9"
```

#### 딕셔너리형(사전구조)
- 딕셔너리변수={"Key1":value1, "key2":value2, ....}

```python
dict1={'이름':'서정동', "성별":"남", "지역":'서울 노원구', "성적": 85}
dict2={"이름":["홍길동","이순신", "박변수"], 
       "성별":["남", "남", "여"], 
       "나이":[35, 66, 28]}
```

```python
print(dict1.keys())
print(dict1.values())
print(dict1["이름"])
```

```python
print(dict2["이름"])
print('이름:'+dict2["이름"][0]+" 성별:"+dict2["성별"][0])
print(f'이름:{dict2["이름"][0]}, 성별:{dict2["성별"][0]}, 나이:{dict2["나이"][0]}')
```

```python
name="서정동"
age=45

print("이름:", name, ", 나이:", age)
print(f'이름:{name}, 성별:{age}')
```
