```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url="https://www.naver.com/"
nav_html=requests.get(url)

soup=bs(nav_html.content, "html.parser")
ul_soup=soup.find('ul', class_="list_nav NM_FAVORITE_LIST")
print(ul_soup)
```

```python
li_soup=ul_soup.find_all('li')
for li in li_soup:
    print(li.text)
```

### 네이버 뉴스 검색 크롤링

```python
import sys

### 뉴스 기사 정보 가져와 파씽 처리
url="https://search.naver.com/search.naver?where=news&sm=tab_jum&query=코로나&start=21"

nav_html=requests.get(url)
if nav_html.status_code == 200:
    nav_soup=bs(nav_html.text, "html.parser")
else:
    sys.exit(0)
```

```python
### 첫페이지 뉴스 제목 가져오기
list_news=nav_soup.find('ul', class_="list_news")
prn_soup=list_news.find_all("a", class_="news_tit")

for a in prn_soup:
    print(a.text)
```

```python
### 첫페이지 요약 내용 져오기
prn_soup=list_news.find_all("a", class_="api_txt_lines dsc_txt_wrap")
for a in prn_soup:
    print(a.text)
```

```python
### 뉴스 제목과 내용 두가지 모두 출력
news_soup=nav_soup.find_all('div', class_="news_area")

for news in news_soup:
    print("[제목]", news.find("a", class_="news_tit").text)
    print("[내용]", news.find("a", class_="api_txt_lines dsc_txt_wrap").get_text())
```

```python
### 뉴스 제목과 내용 두가지 모두 출력
news_soup=nav_soup.find_all('div', class_="news_area")
news_list=[]
for news in news_soup:
    tit=news.find("a", class_="news_tit").text
    body=news.find("a", class_="api_txt_lines dsc_txt_wrap").get_text()
    news_list.append({"제목":tit, "내용":body})
df=pd.DataFrame(news_list)
df
```

```python
### 뉴스 기사 정보 가져와 파씽 처리
def url_parser(sta_page):
    url="https://search.naver.com/search.naver?where=news&sm=tab_jum&query=코로나&start="+str(sta_page)

    nav_html=requests.get(url)
    if nav_html.status_code == 200:
        return bs(nav_html.text, "html.parser")
    else:
        sys.exit(0)
```

```python
news_list=[]
cnt=0
for sta_page in range(1, 32, 10):
    nav_soup=url_parser(sta_page)
    cnt += 1
    
    ### 뉴스 제목과 내용 두가지 모두 출력
    news_soup=nav_soup.find_all('div', class_="news_area")
    for news in news_soup:
        tit=news.find("a", class_="news_tit").text
        body=news.find("a", class_="api_txt_lines dsc_txt_wrap").get_text()
        news_list.append({"페이지번호":cnt, "제목":tit, "내용":body})

df=pd.DataFrame(news_list)
df
```
