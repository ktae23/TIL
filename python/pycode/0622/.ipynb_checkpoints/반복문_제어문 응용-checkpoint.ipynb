```python
from PIL import Image
import pandas as pd
fileName=["pic1", "pic2", "pic3"]
prnData=[]

for fn in fileName:
    tmp=Image.open('./picture/'+fn+'.jpg')
    fName=fn+'.jpg'
    fm=tmp.format
    fw=tmp.width
    fh=tmp.height
    prnData.append([fName, fm, fw, fh])

    tmpRes=tmp.resize((300,300))
    tmpRes.save('./data/'+fn+"_300.jpg")    
    
for prn in prnData:
    print(f'파일명:{prn[0]},\t확장자:{prn[1]},\t너비:{prn[2]},\t높이:{prn[3]}')
    print("   ")

print("+=======================+")

df=pd.DataFrame(prnData, columns=["파일명", "확장자","너비","높이"])
df.to_csv('./파일정보.csv', encoding="cp949")
display(df)

```

```python
from PIL import Image
import os
import pandas as pd

fileDir = './picture/'
fileName=os.listdir(fileDir)
prnData=[]

for fn in fileName:
    if(fn[-3:] == 'jpg'):
        tmp=Image.open(fileDir + fn)
        fm=tmp.format
        fw=tmp.width
        fh=tmp.height
        prnData.append([fName, fm, fw, fh])
        
        tmpRes=tmp.resize((300,300))
        tmpRes.save('./data/'+fn[:-4]+"_300.jpg")    
    
df=pd.DataFrame(prnData, columns=["파일명", "확장자","너비","높이"])
df.to_csv('./파일정보.csv', encoding="cp949")
#display(df)

print("+++++++++++++++++++++++++++++++++++")

fileDir_300 = './data/'
fileName_300=os.listdir(fileDir_300)
prnData_300=[]

for fn in fileName_300:
    if(fn[-3:] == 'jpg'):
        tmp=Image.open(fileDir_300 + fn)
        fm=tmp.format
        fw=tmp.width
        fh=tmp.height
        prnData_300.append([fn, fm, fw, fh])

        
df_300=pd.DataFrame(prnData_300, columns=["파일명", "확장자","너비","높이"])
df_300.to_csv('./리사이즈파일정보.csv', encoding="cp949")
display(df_300)

        
```

```python
txt='코로나19 탓에 손도 열심히 씻고, 마스크를 착용하고, 외출도 줄었다.'
lst=[]
tmp=""
count=0

for i in range(len(txt)):
    if txt[i] == " ":
        count += 1
        lst.append(tmp)
        tmp=""
    tmp = tmp + txt[i]
lst.append(tmp)

print(lst)
print("공백 수: ",count)
        
```
