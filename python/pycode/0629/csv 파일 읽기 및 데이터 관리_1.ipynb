### CSV 형식의 파일 읽어오기
- csv 모듈 이용, open("파일경로 및 파일명", 오픈형식, encoding='cp949')
- 오픈형식: r=읽기전용, w=파일생성 및 쓰기, a=기존파일에 추가
- encoding => cp949: 마이크로소프트에서 한글처리 방법(아스키코드), utf-8: 범용(유니코드)
- pandas 모듈 이용

```python
import csv

f=open('./data/seoul.csv', 'r', encoding='cp949')
data=csv.reader(f, delimiter=",")
print(data)
f.close()
```

```python
f=open('./data/seoul.csv', 'r', encoding='cp949')
data=csv.reader(f, delimiter=",")

header=next(data)   # 첫행(제목행)을 제목행으로 처리
print(header)
print()
for row in data:
    print(row)
    
f.close()
```

### 미션: 최고기온(℃)이 가장 높은 날짜와 최고기온(℃) 출력

```python
f=open('./data/seoul.csv', 'r', encoding='cp949')
data=csv.reader(f, delimiter=",")
header=next(data)   # 첫행(제목행)을 제목행으로 처리

temp_day=""
temp_Max=0

for row in data:
    if row[4] != "":
        if float(row[4])> temp_Max:
            temp_day=row[2]
            temp_Max=float(row[4])
    
f.close()

print(temp_day, temp_Max)
```

### [미션] 자신의 생일과 일치하는 날짜에 대한 날짜/평균기온/최고기온/최저기온(℃) 출력
- 생일일치: 월, 일 두개가 일치하면 나머지 데이터 리스트 변수에 리스트형으로 추가
- 최종 리스트 변수에 추가된 데이터 화면에 출력

```python
f=open('./data/seoul.csv','r', encoding='cp949')
data=csv.reader(f, delimiter=",")

header=next(data) # 첫행(제목행)을 제목행으로 처리
#print(header)

b_lst=["",0,0,0]
B_lst=[]

for row in data:
    if row[2] != "":
        if int(row[2][:4]) >= 1995:
            if row[2][-5:] == '07-23':
                b_lst[0]=row[2]
                b_lst[1]=row[3]
                b_lst[2]=row[4]
                b_lst[3]=row[6]
                B_lst.append(b_lst)
                print(b_lst)
                print(B_lst)
            
f.close()

#B_lst
```

```python
f=open('./data/seoul.csv','r',encoding='cp949')
data=csv.reader(f, delimiter=',')
header=next(data)

birth_list=[]

for row in data:
    BIRTH=['',0,0,0]
    if row[2] != '':
        if int(row[2][:4]) > 1992:
            if row[2][-5:] == "11-07":
                BIRTH[0] = row[2]
                BIRTH[1] = row[3]
                BIRTH[2] = row[4]
                BIRTH[3] = row[6]
            birth_list.append(BIRTH)

f.close

birth_list
```

```python
f=open('./data/seoul.csv','r',encoding='cp949')
data=csv.reader(f, delimiter=',')
header=next(data)

birth_list=[]

for row in data:
    if row[2] != '':
        if int(row[2][:4]) > 1992:
            if row[2][-5:] == "11-07":
                birth_list.append([row[2],row[3],row[4],row[6]])

f.close

birth_list
```
